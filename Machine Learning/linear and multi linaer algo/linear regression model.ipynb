```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
```

```python
#reading of data set
df = pd.read_csv('Housing.csv')
```

```python
df
```

```python
print(df.shape)
df.info()
df.isnull()
```

```python
x=df['garagepl'].values
y=df['price'].values
```

```python
# x
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)


num=0
den=0

for i in range(len(x)):    
 num += (x[i]-mean_x)*(y[i]-mean_y)
 den += (x[i]-mean_x)**2
    
m=num/den    
c=(mean_y)-(m*mean_x)
np.max(x)
```

```python
print(m,c)
```

```python
max_x = np.max(x)
min_x = np.min(x)

X=np.linspace(min_x,max_x,100)
Y=m*X+c

#grap of line
mp.plot(X,Y,color='red',label='line graph')
```

```python
#scater plot graph
mp.scatter(x,y,color='blue',label='scatter plot')
```

```python

#grap of line
mp.plot(X,Y,color='red',label='line graph')
#scater plot graph
mp.scatter(x,y,color='blue',label='scatter plot')
#labels
mp.xlabel('house price')
mp.ylabel('bedrooms')
mp.legend()
mp.show()
```

```python
nume=0
deno=0

#finding r2 valuse
for i in range(len(x)):
    yp=m*x[i]+c
    nume+=(yp-mean_y)**2
    deno+=(y[i]-mean_y)**2
    
r2=nume/deno
```

```python
print(r2)
```
